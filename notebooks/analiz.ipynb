## Satış Analizi Bölümü

```python
import pandas as pd
import seaborn as sns

# Seaborn kütüphanesinin içindeki 1000+ satırlık 'diamonds' (elmas) veri setini çekelim
# Bu veri seti fiyat, kesim kalitesi, renk ve berraklık gibi gerçek veriler içerir.
df_heavy = sns.load_dataset('diamonds')

# Verinin ilk 10 satırına bakalım
print(f"Toplam Satır Sayısı: {len(df_heavy)}")
df_heavy.head(10)
```

```python
# Veri setinde boş değer var mı? Hangi sütun ne tipte?
print(df_heavy.info())

# Sayısal sütunların istatistiksel özeti (Ortalama fiyat, en yüksek fiyat vb.)
df_heavy.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
# Dağılım grafiği (Scatter Plot)
sns.scatterplot(x='carat', y='price', hue='cut', data=df_heavy, alpha=0.5)
plt.title('Elmas Ağırlığı ve Fiyat İlişkisi (Kesim Kalitesine Göre)')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Veriyi oku
netflix_data = pd.read_csv('../data/netflix.csv')

# 2. Verinin genel yapısına bak (Hangi sütunlar var?)
print(netflix_data.columns)

# 3. Netflix'te Film mi çok yoksa Dizi mi? Hemen görelim:
plt.figure(figsize=(8, 5))
sns.countplot(x='type', data=netflix_data, palette='viridis')
plt.title('Netflix İçerik Türü Dağılımı')
plt.show()
```

```python
# Netflix'teki Türkiye (Turkey) içeriklerini bulalım
tr_data = netflix_data[netflix_data['country'] == 'Turkey']

# Bunları bir grafik ile yıllara göre görelim
plt.figure(figsize=(12, 6))
sns.histplot(data=tr_data, x='release_year', bins=20, kde=True, color='red')
plt.title('Netflix Türkiye İçeriklerinin Yıllara Göre Dağılımı')
plt.xlabel('Yayın Yılı')
plt.ylabel('İçerik Sayısı')
plt.show()
```
