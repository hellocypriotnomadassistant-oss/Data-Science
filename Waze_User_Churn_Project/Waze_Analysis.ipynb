```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("waze_data.csv")

# Display the first 5 rows of the dataframe
print("First 5 rows of the dataset:")
display(df.head())

# General information about the dataset
print("\nDataset Information:")
df.info()
```

```python
# Set the visual style
sns.set_theme(style="whitegrid")

# Create a bar chart for user status (Churn vs Retained)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='label', palette='magma')

# Adding labels and title
plt.title('Distribution of User Churn vs. Retained', fontsize=14)
plt.xlabel('User Status', fontsize=12)
plt.ylabel('Number of Users', fontsize=12)

# Display the plot
plt.show()
```

```python
# Descriptive statistics for numerical columns
print("Summary Statistics:")
display(df.describe())
```

```python
# Analyze the relationship between device type and churn status
device_churn = df.groupby(['device', 'label']).size().unstack()
print("\nChurn Counts by Device Type:\n", device_churn)

# Visualization
# Using a stacked bar chart to visualize the ratio of churned vs. retained users per device
device_churn.plot(kind='bar', stacked=True, figsize=(8, 5))
plt.title('Relationship Between Device and Churn Status')
plt.xlabel('Device Type')
plt.ylabel('Number of Users')
plt.xticks(rotation=0) # Keeps device names horizontal for better readability
plt.legend(title='Status')
plt.show()
```

```python
# Analyze the descriptive statistics for 'driven_km_drives' by label
# Comparing mean and median to detect potential skewness
print("\nDriving Distance Statistics (by Label):")
print(df.groupby('label')['driven_km_drives'].agg(['mean', 'median']))

# Visualization of distribution and outliers using a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='label', y='driven_km_drives')
plt.title('Outliers in Distance Traveled (driven_km_drives) by Churn Status')
plt.xlabel('Churn Status')
plt.ylabel('Distance Driven (km)')
plt.show()
```
