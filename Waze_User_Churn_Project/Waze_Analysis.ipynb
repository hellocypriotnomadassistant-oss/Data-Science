```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("waze_data.csv")

# Display the first 5 rows of the dataframe
print("First 5 rows of the dataset:")
display(df.head())

# General information about the dataset
print("\nDataset Information:")
df.info()
```

```python
# Set the visual style
sns.set_theme(style="whitegrid")

# Create a bar chart for user status (Churn vs Retained)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='label', palette='magma')

# Adding labels and title
plt.title('Distribution of User Churn vs. Retained', fontsize=14)
plt.xlabel('User Status', fontsize=12)
plt.ylabel('Number of Users', fontsize=12)

# Display the plot
plt.show()
```

```python
# Descriptive statistics for numerical columns
print("Summary Statistics:")
display(df.describe())
```
